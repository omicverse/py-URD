# py-URD — visualisation R⇄Py parity

`plotDim` coloured by `floodPseudotime` output on Guo qPCR fixture.

## 1. Run R + Py drivers

```python
import subprocess
from pathlib import Path
PORT = Path('..').resolve(); OUT = PORT/'tests'/'plot_compare'
subprocess.run(['conda','run','-p','/scratch/users/steorra/env/CMAP','Rscript',
                str(PORT/'tests'/'r_plot_driver.R'), str(OUT)], check=True, cwd=PORT)
subprocess.run(['python', str(PORT/'tests'/'py_plot_driver.py'), str(OUT)], check=True, cwd=PORT)
print('Renders ready')
```

## 2. plotDim coloured by flood-pseudotime

```python
from PIL import Image
import matplotlib.pyplot as plt
fig, (a1, a2) = plt.subplots(1, 2, figsize=(13, 5))
a1.imshow(Image.open('../tests/plot_compare/R_pseudotime.png')); a1.set_title('R reference'); a1.axis('off')
a2.imshow(Image.open('../tests/plot_compare/Py_pseudotime.png')); a2.set_title('Py port'); a2.axis('off')
plt.suptitle('plotDim(urd, label="pseudotime")'); plt.tight_layout(); plt.show()
```

Same Y-branch topology + same pseudotime ordering (Spearman 0.985 vs R; see MATH.md).
